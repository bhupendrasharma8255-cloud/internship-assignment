```python
matrix = np.random.randint(1, 101, size=(4, 4))

print("\nOriginal 4*4 Matrix:")

print(matrix)

diagonal = np.diag(matrix)
print("\nDiagonal Elements:")
print(diagonal)


modified_matrix = matrix.copy()
modified_matrix[modified_matrix  % 2 == 0] = 0

print("\nMatrix after replacing even numbers with 0:")
print(modified_matrix)

max_index = np.argmax(matrix)
row, col = np.unravel_index(max_index, matrix.shape)

print("\nMaximum Value:", np.max(matrix))
print("Index of Maximum value:", (row, col)
      )
```

<!-- # 4. Using np.linspace() and np.zeros() and np.ones(), create three different 
# arrays and stack them vertically using np.vstack(). Display the final 
# stacked array. -->

```python
arr1 = np.linspace(1, 5, 5)
arr2 = np.zeros(5)
arr3 = np.ones(5)

stacked_array = np.vstack((arr1, arr2, arr3))

print("\nArray from linspace:")
print(arr1)

print("\nArray of zeros:")
print(arr2)

print("\nArray of ones:")
print(arr3)

print("\nFinal Stacked Array:")
print(stacked_array)


```

```python
# Section B – Pandas 
# 5. Create a DataFrame of at least 10 students with columns: Name, Age, 
# Marks, City. Perform the following:  
# • Display the first 5 rows using head() 
# • Show basic statistics using describe() 
# • Filter students who scored more than 75 marks 
# • Sort by Marks in descending order
```

```python
import pandas as pd

# 1. Create a DataFrame of at least 10 students
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack'],
    'Age': [21, 22, 20, 23, 21, 22, 20, 23, 22, 21],
    'Marks': [85, 78, 92, 67, 88, 74, 95, 80, 70, 90],
    'City': ['New York', 'London', 'Paris', 'Berlin', 'Tokyo', 'Sydney', 'Mumbai', 'Toronto', 'Rome', 'Madrid']
}

df = pd.DataFrame(data)

# 2. Display the first 5 rows using head()
print("First 5 rows:")
print(df.head())
print("-" * 30)

# 3. Show basic statistics using describe()
print("Basic statistics:")
print(df.describe())
print("-" * 30)

# 4. Filter students who scored more than 75 marks
print("Students with marks > 75:")
high_scorers = df[df['Marks'] > 75]
print(high_scorers)
print("-" * 30)

# 5. Sort by Marks in descending order
print("Sorted by Marks (Descending):")
sorted_df = df.sort_values(by='Marks', ascending=False)
print(sorted_df)
```

# Marks and City columns. Then:

```python
data = {
    'Name': ['Aman', 'Bhupendra', 'Charlie', 'David', 'Eva', 'Franky', 'lucky', 'gorav', 'chirag', 'Jack'],
    'Age': [20, 21, 19, 22, 20, 23, 21, 22, 20, 19],
    'Marks': [85, 78, 92, 88, 76, 95, 80, 70, 90, 82],
    'City': ['pilani', 'Devroad', 'Chicago', 'jhunjhunu', 'khetri', 
             'Philadelphia', 'loharu', 'bagar', 'chirawa', 'San Jose']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\nMissing Values:")
print(df.isnull())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print("\nAfter Filling Marks:")
print(df)

df = df.dropna(subset=['City'])

print("\nAfter Dropping Missing City:")
print(df)
```

```python


data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile"],
    "Category": ["Electronics", "Electronics", "Electronics",
                 "Electronics", "Electronics"],
    "Sales": [50000, 30000, 20000, 45000, 35000],
    "Region": ["North", "South", "East", "West", "North"]
}

df = pd.DataFrame(data)

# Save CSV (optional)
df.to_csv("sales.csv", index=False)

# Read CSV
df = pd.read_csv("sales.csv")

# Group by Category and calculate total sales
category_sales = df.groupby("Category").agg({"Sales": "sum"})
print("Category-wise Sales:")
print(category_sales)

# Top 3 products by sales
top_products = df.sort_values(by="Sales", ascending=False).head(3)

print("\nTop 3 Products:")
print(top_products)
```

<!-- # 11 Using Seaborn, create a dataset of at least 50 records with columns 
# Age, Salary, Department, Experience. Plot:  
# • A scatter plot of Age vs Salary with hue as Department 
# • A heatmap showing correlation between numeric columns 
# • A bar plot showing average Salary per Department  -->

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset with 50 records
np.random.seed(1)

df = pd.DataFrame({
    'Age': np.random.randint(22, 50, 50),
    'Salary': np.random.randint(25000, 100000, 50),
    'Department': np.random.choice(['HR', 'IT', 'Finance', 'Marketing'], 50),
    'Experience': np.random.randint(1, 20, 50)
})

# Scatter Plot
sns.scatterplot(data=df, x='Age', y='Salary', hue='Department')
plt.title("Age vs Salary")
plt.show()

# Heatmap
corr = df[['Age', 'Salary', 'Experience']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Average Salary per Department
avg_salary = df.groupby('Department')['Salary'].mean().reset_index()

sns.barplot(data=avg_salary, x='Department', y='Salary')
plt.title("Average Salary per Department")
plt.show()
```

<!-- # 12 Create a subplot layout of 2 rows × 2 columns using Matplotlib 
# containing:  
# • Line plot 
# • Scatter plot 
# • Bar chart 
# • Histogram 
# All using the same dataset. Add a main title using suptitle() -->

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
ax[0, 0].plot(x, y)
ax[0, 0].set_title("Line Plot")

# Scatter Plot
ax[0, 1].scatter(x, y)
ax[0, 1].set_title("Scatter Plot")

# Bar Chart
ax[1, 0].bar(x, y)
ax[1, 0].set_title("Bar Chart")

# Histogram
ax[1, 1].hist(y, bins=5)
ax[1, 1].set_title("Histogram")

fig.suptitle("2x2 Matplotlib Subplots", fontsize=16)

plt.tight_layout()
plt.show()
```
